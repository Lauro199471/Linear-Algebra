```python
import numpy as np
import matplotlib.pyplot as plt
```

# Draw Standard Basis in red
basis = [x y]

```python
b_x = [1 , 0]
b_y = [0 , 1]

V = np.array([b_x , b_y])
origin = [0], [0] # origin point

plt. quiver(*origin, V[:,0], V[:,1], color=['r','r'], scale=10)
plt.show()
```

# Draw new basis in blue

```python
e_x = [1 , 2]
e_y = [-2 , 3]

V = np.array([b_x , b_y , e_x , e_y])
origin = [0], [0] # origin point

plt. quiver(*origin, V[:,0], V[:,1], color=['r','r','b','b'], scale=10)
plt.show()
```

# Draw vector in green in standard basis

```python
v_1 = [3 , 5]

b_x = [5 , 0]
b_y = [0 , 5]

V = np.array([b_x , b_y , e_x , e_y , v_1])
origin = [0], [0] # origin point

plt. quiver(*origin, V[:,0], V[:,1], color=['r','r','b','b','g'], scale=17)
plt.show()
```

# Example 2

```python
b_x = [4 , 0]
b_y = [0 , 4]

V = np.array([b_x , b_y])
origin = [0], [0] # origin point

plt. quiver(*origin, V[:,0], V[:,1], color=['r','r'], scale=15)
plt.show()
```

```python
e_x = [2 , 1]
e_y = [-2 , 4]

V = np.array([b_x , b_y , e_x , e_y])
origin = [0], [0] # origin point

plt. quiver(*origin, V[:,0], V[:,1], color=['r','r','b','b'], scale=15)
plt.show()
```

```python
v_1 = [3 , 4]

V = np.array([b_x , b_y , e_x , e_y , v_1])
origin = [0], [0] # origin point

plt. quiver(*origin, V[:,0], V[:,1], color=['r','r','b','b','g'], scale=17)
plt.show()
```

```python
a_mat = np.column_stack((e_x,e_y))
print(a_mat)
```

```python
aInverse = np.linalg.inv(a_mat);
vE = np.dot(aInverse , v_1)
print(vE)
```

```python
V = np.array([b_x , b_y , e_x , e_y , v_1 , vE])
origin = [0], [0] # origin point

plt.quiver(*origin, V[:,0], V[:,1], color=['r','r','b','b','g','c'], scale=17)
plt.show()
```
