# Main Take Away

```python
import math
def rotate(origin, point, angle):
    """
    Rotate a point counterclockwise by a given angle around a given origin.

    The angle should be given in degrees.
    """
    angle = math.radians(angle)
    ox, oy = origin
    px, py = point

    qx = ox + math.cos(angle) * (px - ox) - math.sin(angle) * (py - oy)
    qy = oy + math.sin(angle) * (px - ox) + math.cos(angle) * (py - oy)
    return qx, qy
```

```python
import random
import time
# Define Origin and random point
origin = [0 , 0]
x = random.randint(1, 100)    # Pick a random number between 1 and 100.
y = random.randint(1, 100)    # Pick a random number between 1 and 100.
point =  [x , y]
```

```python
# ==== Add basis ====
basis_one_x = [3 , 0]
basis_one_y = [0 , 4]

normMatrix = [basis_one_x , basis_one_y]
normMatrix_normalized = preprocessing.normalize(normMatrix)  # After normalization.
b_x = normMatrix_normalized[0]
b_y = normMatrix_normalized[1]
```

```python
# == Normalize Data Point according to basis vector
basis_point = [basis_one_x , basis_one_y , point]
normBasis_Point_normalized = preprocessing.normalize(basis_point)  # After normalization.
```

```python
# === Plot ===
V = np.array([b_x , b_y])
origin = [0], [0] # origin point

# === Make Initial Graph ===
plt.axis((-1.5,1.5,-1.5,1.5))
plt.plot([0,0],[-2,2],lw=1, color = "black" , linestyle='--', dashes=(2, 5)) #length of 5, space of 1)
plt.plot([-2,2], [0,0], lw=1, color = "black" , linestyle='--', dashes=(2, 5)) #length of 5, space of 1)

normBasis_Point_normalized[2] = normBasis_Point_normalized[2] / 3 # scaling purposes
plt.quiver(*origin, V[:,0], V[:,1], color=['r','r'], scale=7)
plt.plot([normBasis_Point_normalized[2][0]], [normBasis_Point_normalized[2][1]], marker='o', markersize=3, color="blue") 

xStr = str(round(normBasis_Point_normalized[2][0],3))
yStr = str(round(normBasis_Point_normalized[2][1],3))
text1 = 'Blue pt - x: ' + xStr+ ' y: ' + yStr
plt.text(-1.25 , 1.25 , text1)
plt.savefig('imageInt.png')   # save the figure to file
plt.show()
plt.clf()
```

# Step 1

```python
# === Rotate Point ===
# == First ==========
print("Step 1: Starting")
point = [normBasis_Point_normalized[2][0] , normBasis_Point_normalized[2][1]]
origin = [0 , 0]
for x in range(0, 361,2):
    xPoint , yPoint = rotate(origin=origin, point=point, angle=x);

    # === Plot ===
    i = str(x)
    i = 'image' + i + '.png'
    plt.axis((-1.5,1.5,-1.5,1.5))
    plt.plot([0,0],[-2,2],lw=1, color = "black" , linestyle='--', dashes=(2, 5)) #length of 5, space of 1)
    plt.plot([-2,2], [0,0], lw=1, color = "black" , linestyle='--', dashes=(2, 5)) #length of 5, space of 1)

    plt.plot([xPoint], [yPoint], marker='o', markersize=3, color="green")    
    plt.quiver(*origin, V[:,0], V[:,1], color=['r','r'], scale=7)
    xStr = str(round(xPoint,3))
    yStr = str(round(yPoint,3))
    text1 = 'Blue pt - x: ' + xStr+ ' y: ' + yStr
    plt.text(-1.25 , 1.25 , text1)
    plt.savefig(i)   # save the figure to file
    plt.clf()

print("Step 1: Done")
```

```python
point = [normBasis_Point_normalized[2][0] , normBasis_Point_normalized[2][1]]
origin = [0 , 0]
for x in range(0, 361,2):
    blueAxis_x_new = rotate(origin=origin, point=blueAxis_x, angle=x);
    blueAxis_y_new = rotate(origin=origin, point=blueAxis_y, angle=x);
    xPoint , yPoint = rotate(origin=origin, point=point, angle=x);

    # === Project Data Pt to blue basis ===
    # Ax = b
    blue_basis = [blueAxis_x_new , blueAxis_y_new]
    spacePoint = [[xPoint] , [yPoint]]
    blue_basis_proj = np.dot(blue_basis , spacePoint)
    
    # === Plot ===
    i = str(x)
    i = 'imageInt_step_1_' + i + '.png'
    W = np.array([blueAxis_x_new , blueAxis_y_new])

    plt.axis((-1.5,1.5,-1.5,1.5))
    plt.plot([xPoint], [yPoint], marker='o', markersize=3, color="green")  
    plt.plot([0,0],[-2,2],lw=1, color = "black" , linestyle='--', dashes=(2, 5)) #length of 5, space of 1)
    plt.plot([-2,2], [0,0], lw=1, color = "black" , linestyle='--', dashes=(2, 5)) #length of 5, space of 1)
    plt.quiver(*origin, W[:,0], W[:,1], color=['b','b'], scale=7)
    plt.quiver(*origin, V[:,0], V[:,1], color=['r','r'], scale=7)

    xStr_red = str(round(xPoint,3))
    yStr_red = str(round(yPoint,3))
    text1 = 'Blue pt(red basis) - x: ' + xStr_red + ' y: ' + yStr_red
    plt.text(-1.25 , 1.25 , text1)

    xStr_blue = str(round(blue_basis_proj[0][0],3))
    yStr_blue = str(round(blue_basis_proj[1][0],3))
    text2 = 'Blue pt(blue basis) - x: ' + xStr_blue + ' y: ' + yStr_blue
    plt.text(-1.25 , -1.25 , text2)

    plt.savefig(i)   # save the figure to file
    plt.clf()
    
print("Step 2: Done")    
```
